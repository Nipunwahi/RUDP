```python
import numpy as np
```

```python
def plot(X,Y,xlabel,ylabel,title,name):
    import numpy as np 
    import matplotlib.pyplot as plt 
    fig = plt.figure()
    fig.suptitle(title)
    plt.plot(X,Y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    fig.savefig(name)
```

```python
X = [0,10,20,40,60,80,90]
Y = [0.003,1.004,2.347,7.001,16,44,76]
plot(X,Y,"LOSS PERCENTAGE","TIME TAKEN in seconds","TIME VS LOSS","timevsloss.jpg")
```

```python
X = [10,20,40,60,80,90]
Y = [1.004,2.347,7.001,16,44,76]
Y = np.array(Y)
Y = 14000/Y
plot(X,Y,"LOSS PERCENTAGE","THROUGHPUT in bytes/seconds","THROUGHPUT VS LOSS","throughvsloss.jpg")
```
